#**Correlation Matrix**

An interesting way to identify relationships between variables visually is by using a `correlation matrix.`

It indicates the degree of correlation between variables, demonstrating which ones have the most impact on each other.

Therefore, the lower the number, the lower the correlation between the variables, and the higher the number from `0 to 1`, the higher the level of correlation and impact between the variables.


```python
correlation_matrix = df.corr()
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, annot_kws={"size": 9})

plt.show()
```

Our target variable is "fetal health", so it is interesting to identify the variables that have the greatest impact on defining the values ​​of this classification.

However, when analyzing the other variables and their correlations, it is interesting to note that even in the variables with the highest correlation, there is not a very high degree of correlation between them.

For example, we identified the three highest correlation values ​​with the variable "fetal health";

```python
shades =["#f7b2b0","#c98ea6","#8f7198","#50587f", "#003f5c"]
plt.figure(figsize=(20,10))
sns.boxenplot(data = df,palette = shades)
plt.xticks(rotation=90)
plt.show()
```

In this graph we can have an overview of the situation of each variable.

# **Auto Machine Learning Model - Classification**

`Classification with Auto Machine Learning` is an advanced approach that uses automated machine learning algorithms such as `PyCaret` to develop classification models efficiently and effectively. The main goal is to simplify the process of building classification models, allowing data scientists and machine learning professionals to **save time and effort.**

<left><img width="70%" src="https://images.unsplash.com/photo-1595547131329-3bc81af45c47?auto=format&fit=crop&q=80&w=1974&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D"></left>

Auto Machine Learning automates tasks such as `feature selection`, `hyperparameter tuning`, and `cross-validation`, optimizing model performance. It works through an iterative approach, testing various algorithms and configurations to find the most accurate model. This is particularly useful in scenarios where there is a lot of data and variables to consider.

The goals of classification with Auto Machine Learning include creating accurate classification models, reducing the time required to develop these models, and being able to handle complex data sets. This approach allows data scientists to focus on more strategic tasks, such as interpreting results and making data-driven decisions.

```python
data = df
```

```python
test = data.sample(frac=0.10)
train = data.drop(test.index)

test.reset_index(inplace=True, drop=True)
train.reset_index(inplace=True, drop=True)
```

```python
print(test.shape)
print(train.shape)
```

```python
clf = setup (data = train,
             target = "fetal_health",
             normalize = True,
             transformation = True,
             remove_multicollinearity = True, multicollinearity_threshold = 0.95,
             log_experiment = True,
             experiment_name = "fetal_classification")
```

Below, we will compare and select the best model, with the best results and we will filter by the `"Accuracy"` parameter.

This way, even though there are several possible techniques, we were able to identify the one that best suits our problem and the one that is most likely to **provide better results.**

```python
best = compare_models(sort="Accuracy")
```

```python
print (best)
```

As we can see, the best model (by accuracy) identified was `"Gradient Booster Classifier"` (gbc).

And just above we can check the parameters for this model.

# **Instantiating the Model**

```python
gbc = create_model("gbc")
```

```python
print(gbc)
```

#**Hyperparameter Tuning**

Here we continue to perform hyperparameter tuning to `fine-tune the model.`

This way we can adapt it in the best way to generate more accurate results.

With PyCaret we have the advantage of carrying out this entire process in an `automated way.`

```python
tuned_gbc = tune_model(gbc, optimize = "Accuracy")
```

```python
print(tuned_gbc)
```

In the same way that we can visualize the initially chosen parameters. Now after tuning we can visualize and compare the changes.

In an automated way, the model understood that this way it will have the best results through this organization of parameters.

#**Plot the Model**

With graphical visualizations we can identify interesting patterns and observation points.

To do this, we also use `"plot_model"` in an automated way and define what viewing time should be done.

Below are some important graphs.


```python
plot_model(tuned_gbc, plot = "confusion_matrix")
```

```python
plot_model(tuned_gbc, plot="feature")
```

This is a very important graphical visualization.

Through this we can identify in descending order which variables had the `most weight in defining the model.`

And not just ordering each one of them, but demonstrating `on the blue line`, the level of importance of each one for training the model.

In this case we can identify that the most important variable was `"abnormal_short_term_variability"`

```python
evaluate_model(tuned_gbc)
```

#**Finalize the model**

With all the previous steps done, we can then perform predictions on the test set.

```python
predict_model(tuned_gbc);
```

Using the `finalize_model()` function, the entire dataset, including the part that had `previously been separated` for testing, is used for training.

Therefore, if the model is used to make predictions on the test set separated by PyCaret after the finalize function has been used, the results will be misleading because you will be making predictions on data **already seen** by the model.

To demonstrate this, I will use the finished model with `predict_model()` to compare the results with those we had in the previous part.

```python
final_gbc = finalize_model(tuned_gbc)
```

```python
print(final_gbc)
```

```python
test_preds = predict_model(final_gbc, data=test)
```

```python
test_preds.head()
```
