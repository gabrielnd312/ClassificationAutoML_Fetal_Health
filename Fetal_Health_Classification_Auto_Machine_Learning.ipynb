Our objective is to completely analyze the database, understand the variables and create a Classification model that indicates the fetus' health level.

To achieve this goal, we will utilize the `Classification` technique within the context of `Automated Machine Learning (Auto ML)`.

Furthermore, we will employ the PyCaret library, which is a powerful tool to streamline and optimize our model-building process.

`Automated Machine Learning (Auto ML)` is an approach designed to make the development of machine learning models more efficient and accessible.

It automates tasks such as algorithm selection, hyperparameter tuning, and model evaluation, allowing data scientists to focus on interpreting results and creating solutions. This approach is especially valuable when dealing with large datasets like that of Health Insurance, where we aim to gain meaningful insights effectively.

Auto Machine Learning Model - Classification

```python
data = df
```

```python
test = data.sample(frac=0.10)
train = data.drop(test.index)

test.reset_index(inplace=True, drop=True)
train.reset_index(inplace=True, drop=True)
```

```python
print(test.shape)
print(train.shape)
```

```python
clf = setup (data = train,
             target = "fetal_health",
             normalize = True,
             transformation = True,
             remove_multicollinearity = True, multicollinearity_threshold = 0.95,
             log_experiment = True,
             experiment_name = "fetal_classification")
```

```python
best = compare_models(sort="Accuracy")
```

```python
print (best)
```

# **Instantiating the Model**

```python
lgb = create_model("lightgbm")
```

```python
print(lgb)
```

#**Hyperparameter Tuning**

```python
tuned_lgb = tune_model(lgb, optimize = "Accuracy")
```

```python
print(tuned_lgb)
```

#**Plot the Model**

```python
plot_model(tuned_lgb, plot = "confusion_matrix")
```

```python
plot_model(tuned_lgb, plot="feature")
```

```python
evaluate_model(tuned_lgb)
```

#**Finalize the model**


```python
predict_model(tuned_lgb);
```

```python
final_lgb = finalize_model(tuned_lgb)
```

```python
print(final_lgb)
```
